```python
import requests
from lxml import html
```

```python
url = 'https://www.ezday.co.kr/bbs/srh_board.html'
params = {
    'maniaID':'',
    'q_id_info':743,
    'q_left_menu':'',
    'srh[scal]':10,
    'srh[sort]':'date',
    'srh[date]':'all',
    'srh[page]':1, #pagination
}
```

```python
res = requests.get(url, params=params)
res
```

```python
res.text
```

```python
root = html.fromstring(res.text)
```

```python
urls = root.xpath('//div[@class="cont_mania"]/strong/a/@href')
urls
```

```python
res = requests.get('https://www.ezday.co.kr' + urls[0])
res.encoding='EUC-KR'
```

```python
root = html.fromstring(res.text)
```
