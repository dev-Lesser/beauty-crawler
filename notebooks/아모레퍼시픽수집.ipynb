```python
import requests
from lxml import html
```

```python
# 스킨케어 제품 리스트
url = 'https://www.amoremall.com/kr/ko/display/prodListEx/CTG001'
data = {
#    flag: 
# attr: 
'prodSort': 'Bestselling7', # 신상품순 NewProd
'offset': 0 ,# 0, 20 , 40 으로 20씩
'limit': 20,
'includeFilters': 'true',
'displayCateDepth': '99',
'displayCate': '',
'brandList': '',
'priceRange': '',
'prefixMdPicked': 'true',
}
```

```python
# 이 header 의 pair 가 반드시 필요
# Session 값은 cookie 로 가져오는데 암호화하는 x-csrf-token 이 필요함
# default 로 일단 static 변수로 활용하고 다음날 확인
headers={
    'X-CSRF-TOKEN': 'd9203d45-af6f-4b12-be00-c13bb70dae40',
    'Cookie': 'SESSION=21eb9386-57a3-4749-9895-35d420770369;'
}
res = requests.post(url, data=data, headers=headers)
res
```

```python
item_number = [i.get('onlineProdSn') for i in res.json()['filterableOnlineProdList']['list']]
item_number
```
