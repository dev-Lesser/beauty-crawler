```python
import requests
from lxml import html
```

```python
url = 'https://www.naturecollection.com/mall/product/ajax/getProdList.jsp'
```

```python
data = {
    'cate_seq': 7,
    'page_size': 24,
    'sort_type': 'BEST', # best 상품순
    'page': 1, # pagenation
    'pagesize': 24,
}
res = requests.post(url ,data=data)
```

```python
res.text
```

```python
root = html.fromstring(res.text)
```

#### 이미지 url

```python
images = root.xpath('//div[@class="img"]/img/@src')
images[:5]
```

#### 상품 이름

```python
[i.text_content().strip() for i in root.xpath('//p[contains(@class,"prdt_name_1")]')]
```

```python
price_root = root.xpath('//li')
```

### sale가랑 정상가

```python
for path in price_root:
    r = [i.text_content().strip() for i in path.xpath('./span[@class="prdt_price_1"]')]
    sale =  [i.text_content().strip() for i in path.xpath('./span[@class="prdt_price_9"]')]
    if r:
        print(r[0])
    if sale:
        print(sale[0])
```
