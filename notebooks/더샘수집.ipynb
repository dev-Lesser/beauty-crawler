```python
import requests
from lxml import html
```

```python
url = 'https://healivemall.com/product/list.html'
```

```python
params = {
    'cate_no':84, # 스킨케어
    'page':1
}
headers= {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36'
}
```

```python
res = requests.get(url, params=params, headers=headers)
```

```python
res
```

```python
res.text
```

```python
root = html.fromstring(res.text)
```

### 상품 url 

```python
root.xpath('//div[@class="product-image"]/a/@href')
```

### 상품 이미지

```python
root.xpath('//div[@class="product-image"]/a/img/@src')
```

#### 상품이름

```python
root.xpath('//div[@class="prod_name"]/span/text()')
```

#### 세일가 & 정상가

```python
[i.text_content().strip() for i in root.xpath('//div[@class="price"]/p[@class="sale"]')],[i.text_content().strip() for i in root.xpath('//p[@class="won"]')]
```
